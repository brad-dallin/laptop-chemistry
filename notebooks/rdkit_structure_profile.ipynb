```python
# Import modules
import rdkit
import pandas as pd
from notebook_utils import draw_molecule_acs1996
# Expand to see all columns
pd.set_option('display.max_columns', None)

# Suppress RDKit Output
rdkit.RDLogger.DisableLog('rdApp.*')

# Print versions
print(f"Pandas Version: {pd.__version__}")
print(f"RDKit Version: {rdkit.__version__}")
```

```python
smiles = "Nc1ccc(S(=O)(=O)Nc2ncccn2)cc1"
mol = rdkit.Chem.MolFromSmiles(smiles)

# Find all nitrogen atoms in the molecule
nitrogen_atoms = [atom.GetIdx() for atom in mol.GetAtoms() if atom.GetAtomicNum() == 7]

# Highlight nitrogen atoms in blue
img = draw_molecule_acs1996(mol, 
                  legend="Sulfadiazine", 
                  highlight_atoms=nitrogen_atoms,
                  highlight_color=(1, 1, 0.7))
display(img)
```

```python
smiles = "Nc1ccc(S(=O)(=O)Nc2ncccn2)cc1"
mol = rdkit.Chem.MolFromSmiles(smiles)

# Find all aromatic bonds in the molecule
aromatic_bonds = [bond.GetIdx() for bond in mol.GetBonds() if bond.GetIsAromatic()]

# Highlight aromatic bonds in purple
img = draw_molecule_acs1996(mol, 
                  legend="Sulfadiazine", 
                  highlight_bonds=aromatic_bonds,
                  highlight_bond_color=(0.8, 0.4, 0.8))  # Purple color
display(img)
```
